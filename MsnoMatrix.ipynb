```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/employees.csv')
print(df.head())
print(df.dtypes)

df['Gender'] = df['Gender'].astype('category')
print(df.dtypes)

print('----------------------------')

df.isnull().sum()
df.fillna(method='ffill', inplace=True)
import missingno as msno
msno.matrix(df)

df_encoded = pd.get_dummies(df, columns=['Gender'], dtype=np.int64)
df_encoded.head()
# df.isnull().sum()
```
